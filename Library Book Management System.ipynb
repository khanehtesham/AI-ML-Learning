```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.is_borrowed = False
        self.borrower = None

    def info(self):
        status = f"Borrowed by {self.borrower}" if self.is_borrowed else "Available"
        print(f"{self.title} by {self.author} - {status}")  

    def borrow(self, user):
        if not self.is_borrowed:
            self.is_borrowed = True
            self.borrower = user
            print(f"'{self.title}' has been borrowed by {user}.")
        else:
            print(f"Sorry,'{self.title}' is already borrowed by {self.borrower}.")
           

    def return_book(self):
         if self.is_borrowed:
            print(f"'{self.title}' has been returned by {self.borrower}.")
            self.is_borrowed = False
            self.borrower = None
         else:
            print(f"'{self.title}' is not currently borrowed.")



    ####### Libarary Class ####### 
    class Library:
        def __init__(self):
            self.books=[]

        def add_book(self,book):
            self.books.append(book)
            print(f" Book '{book.title} added to the library.")

        def display_books(self):
            if not self.books:
                print("No books in the Library")
            else:
                print("\nüìò Library Catalog:")
                for book in self.books:
                    book.info()

        def find_book(self, title):
            for book in self.books:
                if book.title.lower() == title.lower():
                    return book
            return None

        def borrow_book(self, title, user):
            book = self.find_book(title)
            if book:
                book.borrow(user)
            else:
                print(f"‚ùå Book '{title}' not found in library.")

        def return_book(self, title):
                book = self.find_book(title)
                if book:
                  book.return_book()
                else:
                  print(f"‚ùå Book '{title}' not found in library.")        


    book = Book("Atomic Habits", "James Clear")
    book.info()
    book.borrow("Alice")
    book.info()
    book.return_book()
    book.info()

    
    lib = Library()
    b1 = Book("The Hobbit", "J.R.R. Tolkien")
    b2 = Book("The Great Gatsby", "F. Scott Fitzgerald")

    # Add books to the library
    lib.add_book(b1)
    lib.add_book(b2)

    # Display all books
    lib.display_books()
    
    # Borrow a book
    lib.borrow_book("The Hobbit", "John")
    
    # Try to borrow the same book again
    lib.borrow_book("The Hobbit", "Emily")
    
    # Return the book
    lib.return_book("The Hobbit")
    
    # Try to return again
    lib.return_book("The Hobbit")
    
   

 
    
    
```
